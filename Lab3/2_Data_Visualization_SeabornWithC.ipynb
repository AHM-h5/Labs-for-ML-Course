# Seaborn Overview

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.




## Distribution Plots

Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

* distplot
* jointplot
* pairplot
* rugplot
* kdeplot

## Imports

```python
import seaborn as sns # Imports the seaborn library and assigns it the alias `sns`.
%matplotlib inline # A magic command to display matplotlib plots directly in the notebook.
```

## Data
Seaborn comes with built-in data sets!

```python
tips = sns.load_dataset('tips') # Loads the 'tips' dataset from Seaborn's built-in datasets.
```

```python
tips.head() # Displays the first 5 rows of the `tips` DataFrame.
```

## distplot

The distplot shows the distribution of a univariate set of observations.

```python
sns.distplot(tips['total_bill']) # Creates a distribution plot (histogram with KDE) of the 'total_bill' column from the `tips` DataFrame.
# Safe to ignore warnings
```

To remove the kde layer and just have the histogram use:

```python
sns.distplot(tips['total_bill'],kde=False,bins=30) # Creates a distribution plot (histogram only, no KDE) of the 'total_bill' column with 30 bins.
```

## jointplot

jointplot() allows you to basically match up two distplots for bivariate data. With your choice of what **kind** parameter to compare with:
* “scatter”
* “reg”
* “resid”
* “kde”
* “hex”

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter') # Creates a joint plot showing the relationship between 'total_bill' and 'tip' as a scatter plot.
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex') # Creates a joint plot showing the relationship between 'total_bill' and 'tip' as a hexagonal binning plot.
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg') # Creates a joint plot showing the relationship between 'total_bill' and 'tip' with a linear regression model fit.
```

## pairplot

pairplot will plot pairwise relationships across an entire dataframe (for the numerical columns) and supports a color hue argument (for categorical columns).

```python
sns.pairplot(tips) # Creates pairwise relationships across an entire DataFrame for numerical columns.
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm') # Creates pairwise relationships across a DataFrame, with points colored by the 'sex' column and using the 'coolwarm' color palette.
```

## rugplot

rugplots are actually a very simple concept, they just draw a dash mark for every point on a univariate distribution. They are the building block of a KDE plot:

```python
sns.rugplot(tips['total_bill']) # Draws a rug plot (dashes for each data point) for the 'total_bill' column.
```

## kdeplot

kdeplots are [Kernel Density Estimation plots](http://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth). These KDE plots replace every single observation with a Gaussian (Normal) distribution centered around that value. For example:

```python
# Don't worry about understanding this code!
# It's just for the diagram below
import numpy as np # Imports the numpy library and assigns it the alias `np`.
import matplotlib.pyplot as plt # Imports the matplotlib.pyplot module and assigns it the alias `plt`.
from scipy import stats # Imports the `stats` module from the `scipy` library.

#Create dataset
dataset = np.random.randn(25) # Generates an array of 25 random numbers from a standard normal distribution.

# Create another rugplot
sns.rugplot(dataset); # Draws a rug plot for the `dataset`.

# Set up the x-axis for the plot
x_min = dataset.min() - 2 # Calculates the minimum value of the dataset and subtracts 2.
x_max = dataset.max() + 2 # Calculates the maximum value of the dataset and adds 2.

# 100 equally spaced points from x_min to x_max
x_axis = np.linspace(x_min,x_max,100) # Creates an array of 100 evenly spaced numbers over the specified range.

# Set up the bandwidth, for info on this:
url = 'http://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth' # Defines a URL for reference.

bandwidth = ((4*dataset.std()**5)/(3*len(dataset)))**.2 # Calculates the optimal bandwidth for Kernel Density Estimation.


# Create an empty kernel list
kernel_list = [] # Initializes an empty list to store kernel functions.

# Plot each basis function
for data_point in dataset: # Loops through each data point in the dataset.

    # Create a kernel for each point and append to list
    kernel = stats.norm(data_point,bandwidth).pdf(x_axis) # Calculates the probability density function of a normal distribution centered at `data_point`.
    kernel_list.append(kernel) # Appends the calculated kernel to the `kernel_list`.

    #Scale for plotting
    kernel = kernel / kernel.max() # Normalizes the kernel to its maximum value.
    kernel = kernel * .4 # Scales the kernel for plotting purposes.
    plt.plot(x_axis,kernel,color = 'grey',alpha=0.5) # Plots the scaled kernel function.

plt.ylim(0,1) # Sets the y-axis limits of the plot.
```

```python
# To get the kde plot we can sum these basis functions.

# Plot the sum of the basis function
sum_of_kde = np.sum(kernel_list,axis=0) # Calculates the sum of all kernel functions along the specified axis.

# Plot figure
fig = plt.plot(x_axis,sum_of_kde,color='indianred') # Plots the sum of the basis functions with a specified color.

# Add the initial rugplot
sns.rugplot(dataset,c = 'indianred') # Draws a rug plot for the `dataset` with a specified color.

# Get rid of y-tick marks
plt.yticks([]) # Removes y-axis tick marks.

# Set title
plt.suptitle("Sum of the Basis Functions") # Sets the main title of the plot.
```

So with our tips dataset:

```python
sns.kdeplot(tips['total_bill']) # Creates a Kernel Density Estimate plot for the 'total_bill' column.
sns.rugplot(tips['total_bill']) # Draws a rug plot for the 'total_bill' column.
```

```python
sns.kdeplot(tips['tip']) # Creates a Kernel Density Estimate plot for the 'tip' column.
sns.rugplot(tips['tip']) # Draws a rug plot for the 'tip' column.
```

# Categorical Data Plots

Now let's discuss using seaborn to plot categorical data! There are a few main plot types for this:

* factorplot
* boxplot
* violinplot
* stripplot
* swarmplot
* barplot
* countplot

Let's go through examples of each!

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips') # Loads the 'tips' dataset from Seaborn's built-in datasets.
tips.head() # Displays the first 5 rows of the `tips` DataFrame.
```

## barplot and countplot

These very similar plots allow you to get aggregate data off a categorical feature in your data. **barplot** is a general plot that allows you to aggregate the categorical data based off some function, by default the mean:

```python
sns.barplot(x='sex',y='total_bill',data=tips) # Creates a bar plot showing the mean 'total_bill' for each 'sex' category.
```

```python
import numpy as np # Imports the numpy library and assigns it the alias `np`.
```

You can change the estimator object to your own function, that converts a vector to a scalar:

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std) # Creates a bar plot showing the standard deviation of 'total_bill' for each 'sex' category.
```

### countplot

This is essentially the same as barplot except the estimator is explicitly counting the number of occurrences. Which is why we only pass the x value:

```python
sns.countplot(x='sex',data=tips) # Creates a count plot showing the number of occurrences for each 'sex' category.
```

## boxplot and violinplot

boxplots and violinplots are used to shown the distribution of categorical data. A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range.

```python
sns.boxplot(x="day", y="total_bill", data=tips,palette='rainbow') # Creates a box plot showing the distribution of 'total_bill' for each 'day' category, using the 'rainbow' palette.
```

```python
# Can do entire dataframe with orient='h'
sns.boxplot(data=tips,palette='rainbow',orient='h') # Creates horizontal box plots for all numerical columns in the `tips` DataFrame, using the 'rainbow' palette.
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips, palette="coolwarm") # Creates a box plot showing the distribution of 'total_bill' for each 'day', split by 'smoker' status, using the 'coolwarm' palette.
```

### violinplot
A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared. Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution.

```python
sns.violinplot(x="day", y="total_bill", data=tips,palette='rainbow') # Creates a violin plot showing the distribution of 'total_bill' for each 'day' category, using the 'rainbow' palette.
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',palette='Set1') # Creates a violin plot showing the distribution of 'total_bill' for each 'day', split by 'sex', using the 'Set1' palette.
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',split=True,palette='Set1') # Creates a violin plot showing the distribution of 'total_bill' for each 'day', split by 'sex' and with the violins split, using the 'Set1' palette.
```

## stripplot and swarmplot
The stripplot will draw a scatterplot where one variable is categorical. A strip plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.

The swarmplot is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they don’t overlap. This gives a better representation of the distribution of values, although it does not scale as well to large numbers of observations (both in terms of the ability to show all the points and in terms of the computation needed to arrange them).

```python
sns.stripplot(x="day", y="total_bill", data=tips, palette='rainbow') # Creates a strip plot showing individual data points of 'total_bill' for each 'day', using the 'rainbow' palette.
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True, palette='rainbow') # Creates a strip plot with added jitter to prevent overlapping points, showing 'total_bill' for each 'day', using the 'rainbow' palette.
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1') # Creates a jittered strip plot showing 'total_bill' for each 'day', with points colored by 'sex' and using the 'Set1' palette.
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1',dodge=True) # Creates a jittered strip plot showing 'total_bill' for each 'day', with points colored by 'sex', using the 'Set1' palette, and with points for different hue levels dodged.
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips) # Creates a swarm plot showing the distribution of 'total_bill' for each 'day', with non-overlapping points.
```

### Combining Categorical Plots

```python
sns.violinplot(x="tip", y="day", data=tips,palette='rainbow') # Creates a violin plot showing the distribution of 'tip' for each 'day', using the 'rainbow' palette.
sns.swarmplot(x="tip", y="day", data=tips,color='black',size=3) # Overlays a swarm plot on the violin plot, showing individual data points of 'tip' for each 'day', colored black and with a size of 3.
```

## catplot

factorplot is the most general form of a categorical plot. It can take in a kind parameter to adjust the plot type:

```python
sns.catplot(x='sex',y='total_bill',data=tips,kind='bar') # Creates a categorical plot of kind 'bar', showing the mean 'total_bill' for each 'sex' category.
```

## Heatmap

In order for a heatmap to work properly, your data should already be in a matrix form, the sns.heatmap function basically just colors it in for you. For example:

```python
tips.head() # Displays the first 5 rows of the `tips` DataFrame.
```

```python
# Matrix form for correlation data
tips.corr() # Calculates the pairwise correlation between columns in the `tips` DataFrame.
```

```python
sns.heatmap(tips.corr()) # Creates a heatmap of the correlation matrix of the `tips` DataFrame.
```

```python
sns.heatmap(tips.corr(),cmap='coolwarm',annot=True) # Creates a heatmap of the correlation matrix, using the 'coolwarm' colormap and annotating each cell with its value.
```

Or for the flights data:

```python
flights.pivot_table(values='passengers',index='month',columns='year') # Creates a pivot table from the `flights` DataFrame, with 'month' as index, 'year' as columns, and 'passengers' as values.
```

```python
pvflights = flights.pivot_table(values='passengers',index='month',columns='year') # Creates a pivot table from the `flights` DataFrame, with 'month' as index, 'year' as columns, and 'passengers' as values, and assigns it to `pvflights`.
sns.heatmap(pvflights) # Creates a heatmap of the `pvflights` pivot table.
```

```python
sns.heatmap(pvflights,cmap='magma',linecolor='white',linewidths=1) # Creates a heatmap of `pvflights`, using the 'magma' colormap, white lines between cells, and a linewidth of 1.
```

## clustermap

The clustermap uses hierarchal clustering to produce a clustered version of the heatmap. For example:

```python
sns.clustermap(pvflights) # Creates a hierarchically-clustered heatmap of the `pvflights` pivot table.
```

Notice now how the years and months are no longer in order, instead they are grouped by similarity in value (passenger count). That means we can begin to infer things from this plot, such as August and July being similar (makes sense, since they are both summer travel months)

```python
# More options to get the information a little clearer like normalization
sns.clustermap(pvflights,standard_scale=1) # Creates a hierarchically-clustered heatmap of `pvflights`, with rows standardized by Z-score.
```

# Regression Plots

Seaborn has many built-in capabilities for regression plots, however we won't really discuss regression until the machine learning section of the course, so we will only cover the **lmplot()** function for now.

**lmplot** allows you to display linear models, but it also conveniently allows you to split up those plots based off of features, as well as coloring the hue based off of features.

Let's explore how this works:

```python
import seaborn as sns # Imports the seaborn library and assigns it the alias `sns`.
%matplotlib inline # A magic command to display matplotlib plots directly in the notebook.
```

```python
tips = sns.load_dataset('tips') # Loads the 'tips' dataset from Seaborn's built-in datasets.
```

```python
tips.head() # Displays the first 5 rows of the `tips` DataFrame.
```

## lmplot()

```python
sns.lmplot(x='total_bill',y='tip',data=tips) # Creates a linear model plot showing the relationship between 'total_bill' and 'tip'.
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex') # Creates a linear model plot showing the relationship between 'total_bill' and 'tip', with points colored by 'sex'.
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm') # Creates a linear model plot showing the relationship between 'total_bill' and 'tip', with points colored by 'sex' and using the 'coolwarm' color palette.
```

## Using a Grid

We can add more variable separation through columns and rows with the use of a grid. Just indicate this with the col or row arguments:

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex') # Creates a linear model plot showing the relationship between 'total_bill' and 'tip', with separate plots for each 'sex' category arranged in columns.
```

```python
sns.lmplot(x="total_bill", y="tip", row="sex", col="time",data=tips) # Creates a linear model plot showing the relationship between 'total_bill' and 'tip', with separate plots arranged in rows by 'sex' and in columns by 'time'.
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm') # Creates a linear model plot showing the relationship between 'total_bill' and 'tip', with separate plots for each 'day' arranged in columns, points colored by 'sex', and using the 'coolwarm' palette.
```

## Aspect and Size

Seaborn figures can have their size and aspect ratio adjusted with the **height** and **aspect** parameters:

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm',
          aspect=0.6,height=8) # Creates a linear model plot showing the relationship between 'total_bill' and 'tip', with separate plots for each 'day' arranged in columns, points colored by 'sex', using the 'coolwarm' palette, with an aspect ratio of 0.6 and a height of 8 inches.
```
